```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import statsmodels.api as sm
%matplotlib inline
```

```python
df = pd.read_csv('data/kc_house_data.csv')
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.drop('sqft_basement', axis=1, inplace=True)
```

```python
df.hist(figsize=(10,7))
```

```python
import statsmodels.formula as smf
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python

outcome = 'price'
x_cols = df.drop(['price', 'date', 'id', 'zipcode', 'sqft_above','sqft_lot15'], axis = 1).columns
predictors = '+'.join(x_cols)

f = outcome + '~' + predictors
```

```python
from sklearn.model_selection import train_test_split

data = df.copy()
y = data['price']
X = data.drop(['price', 'date', 'id', 'zipcode', 'sqft_above'], axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

linreg = LinearRegression()
linreg.fit(X_train, y_train)

y_hat_train = linreg.predict(X_train)
y_hat_test = linreg.predict(X_test)
from sklearn.metrics import mean_squared_error

mse_train = mean_squared_error(y_train, y_hat_train)
mse_test = mean_squared_error(y_test, y_hat_test)

print('Train MSE:', mse_train)
print('Test MSE:', mse_test)

print('RMSE Train:', np.sqrt(mse_train))
print('RMSE Test:', np.sqrt(mse_test))

from sklearn.metrics import r2_score

r2_score(y_test, y_hat_test)


residuals = (y_test - y_hat_test)

import statsmodels

statsmodels.graphics.gofplots.qqplot(residuals, line = "r")
```

## Model 2 with zipcode dummies

```python
zip_dummies = pd.get_dummies(df['zipcode'],prefix='zipcode')
```

```python
df_with_zip = pd.concat([df,zip_dummies],axis=1)
```

```python
df_with_zip
```

```python
outcome = 'price'
x_cols = df_with_zip.drop(['price', 'date', 'id', 'zipcode', 'sqft_above','sqft_lot15'], axis = 1).columns
predictors = '+'.join(x_cols)

f = outcome + '~' + predictors
```

```python
model2 = ols(formula=f, data= df_with_zip).fit()
model2.summary()
```

```python
from sklearn.model_selection import train_test_split

data = df_with_zip.copy()
y = data['price']
X = data.drop(['price', 'date', 'id', 'zipcode', 'sqft_above'], axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

linreg = LinearRegression()
linreg.fit(X_train, y_train)

y_hat_train = linreg.predict(X_train)
y_hat_test = linreg.predict(X_test)
from sklearn.metrics import mean_squared_error

mse_train = mean_squared_error(y_train, y_hat_train)
mse_test = mean_squared_error(y_test, y_hat_test)

print('Train MSE:', mse_train)
print('Test MSE:', mse_test)

print('RMSE Train:', np.sqrt(mse_train))
print('RMSE Test:', np.sqrt(mse_test))

from sklearn.metrics import r2_score

r2_score(y_test, y_hat_test)


residuals = (y_test - y_hat_test)

import statsmodels

statsmodels.graphics.gofplots.qqplot(residuals, line = "r")
```

```python
from scipy import stats
dfz = df.drop('date',axis=1)
dfz = dfz[(np.abs(stats.zscore(dfz)) < 3).all(axis=1)]
```
